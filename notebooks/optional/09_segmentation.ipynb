## Introduction

Segmentation assigns each pixel to a labelled region such as a grain, domain, or defect, and is one of the most practically important tasks in SPM image analysis. Unlike spectral clustering (Notebook 02), which groups pixels by their spectrum alone, segmentation methods exploit spatial context — neighbouring pixels are more likely to belong to the same region. Classical approaches include intensity thresholding (e.g. Otsu's method), morphological watershed transforms that flood from local minima, and superpixel algorithms such as SLIC that first merge pixels into compact, perceptually uniform patches. These methods are fast, interpretable, and readily composable with downstream machine-learning steps.
