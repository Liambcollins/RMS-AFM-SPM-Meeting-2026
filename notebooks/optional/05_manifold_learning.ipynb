## Introduction

Hyperspectral SPM data contains one full spectrum per pixel, making the dataset potentially 64–512 dimensional. Manifold learning algorithms such as t-SNE and UMAP compress this high-dimensional space to 2D for visualisation while attempting to preserve neighbourhood structure. The resulting scatter plot can be colour-coded by cluster membership or by spatial location on the original image, revealing chemical or structural domains that are invisible in individual channel maps. These methods complement dimensionality-reduction techniques such as PCA (Notebook 01) by capturing non-linear structure that PCA misses.
