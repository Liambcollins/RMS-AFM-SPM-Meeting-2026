## Introduction

Standard ML treats spectral data as generic feature vectors, ignoring the fact that resonance peaks, force curves, and I–V spectra are governed by well-understood physical models. By embedding known physics — such as a Lorentzian lineshape or Hertzian contact mechanics — directly into the ML pipeline, we constrain the solution space and achieve better parameter extraction from noisy measurements than either pure fitting or pure ML alone. Physics-informed approaches range from custom loss functions that penalise physically inconsistent predictions, to hybrid architectures where a neural network corrects the residuals of an analytic model. These ideas are increasingly important as SPM datasets grow larger and more automated analysis pipelines are needed.

## Section 4 — Discussion: Physics-Informed Approaches in SPM

### Why embed physics into ML for SPM?

SPM measurements are inherently noisy, with signal-to-noise ratios that degrade rapidly away from optimal tip–sample conditions. Physical models (Lorentzian resonance, Hertz contact, tunnelling decay, Kelvin probe response) encapsulate decades of domain knowledge that should not be discarded when applying ML.

### Key strategies

| Strategy | Description | SPM example |
|---|---|---|
| **Physics loss** | Penalise predictions that violate a known equation | Reconstruct spectrum from predicted params; minimise residual |
| **Constrained outputs** | Enforce physical bounds via output activation | Force Q-factor > 0, resonance frequency in band |
| **Hybrid model** | Analytic model + NN residual corrector | Hertz baseline + NN for adhesion corrections |
| **Symbolic regression** | Discover closed-form expressions from data | Rediscover contact-mechanics scaling laws |
| **Data augmentation** | Generate training data from the physical model | Simulate Lorentzians with randomised params |

### Limitations

- If the physical model is wrong or incomplete (e.g. non-Lorentzian lineshape due to coupling), embedding it as a hard constraint can introduce systematic bias.
- Symbolic regression scales poorly to high-dimensional inputs.
- Physics-informed neural networks (PINNs) require automatic differentiation frameworks (PyTorch, JAX) and are non-trivial to train stably.
